```python
import pandas as pd
import numpy as np
file_path = r'D:\Assignments\Assignment One\2\time-series.csv'  
```

```python
df = pd.read_csv(file_path, index_col=0, parse_dates=True)
```

```python
df
```

```python
units = {
    'DK_1_load_MW': 'MW',
    'DK_1_wind_MW': 'MW',
    'DK_1_solar_MW': 'MW',
    'DK_1_price_eur_MWh': '€/MWh'
}

### Task (a): Means, Min, Max ###
stats = df.agg(['mean', 'min', 'max']).T
stats.columns = ['Mean', 'Min', 'Max']

print("=== Task (a): Mean, Min, Max ===")
for col in stats.index:
    print(f"{col}: Mean = {stats.loc[col, 'Mean']:.2f} {units[col]}, "
          f"Min = {stats.loc[col, 'Min']:.2f} {units[col]}, "
          f"Max = {stats.loc[col, 'Max']:.2f} {units[col]}")
```

```python
### Task (b): Normalized Standard Deviations ###
df_normed = df / df.max()
std_devs = df_normed.std()
max_std_col = std_devs.idxmax()


print(f"{max_std_col} has the highest normalized standard deviation: {std_devs[max_std_col]:.4f}")
```

```python
### Task (c): Negative Electricity Prices ###
neg_prices = df[df['DK_1_price_eur_MWh'] < 0]


print(neg_prices[['DK_1_price_eur_MWh']].tail(10))

print(f"\nTotal number of hours with negative prices: {neg_prices.shape[0]}")
```

```python
# Calculate total renewable generation
renewable_gen = df['DK_1_wind_MW'] + df['DK_1_solar_MW']

# Compare with demand
exceeds_demand = renewable_gen > df['DK_1_load_MW']

# Count percentage
percentage = exceeds_demand.sum() / len(df) * 100

print(f"Wind + Solar generation exceeds demand in {percentage:.2f}% of the hours.")
```

```python
correlation = df['DK_1_wind_MW'].corr(df['DK_1_price_eur_MWh'])
print(f"✅ Correlation coefficient between wind generation and price: {correlation:.4f}")
```
